```python
import random

choices = ['rock', 'paper', 'scissors']

cpu_choice = random.choice(choices)

current_turn = 1
max_turn = 3
win_counter = 0
loss_counter = 0
tie_counter = 0

print("Welcome to Rock Paper Scissors!")
print("The idea's simple. Rock beats Scissors, Scissors beats Paper, Paper beats Rock.")
print("Beat the computer within three tries to win!")


while current_turn <= max_turn:
    player_choice = input("Choose your weapon: ").lower()
    if player_choice in choices:
        if player_choice == cpu_choice:
            print("Tie!", end=" ")
            print("Computer chose {}".format(cpu_choice))
            print("Turn {} ended".format(current_turn))
            current_turn +=1
            tie_counter +=1
            cpu_choice = random.choice(choices)
        elif player_choice == 'rock' and cpu_choice == 'scissors' or player_choice == 'paper' and cpu_choice == 'rock' or player_choice == 'scissors' and cpu_choice == 'paper':
            print("You win this turn!", end=" ")
            print("Computer chose {}".format(cpu_choice))
            print("Turn {} ended".format(current_turn))
            current_turn +=1
            win_counter +=1
            cpu_choice = random.choice(choices)
        elif player_choice == 'rock' and cpu_choice == 'paper' or player_choice == 'paper' and cpu_choice == 'scissors' or player_choice == 'scissors' and cpu_choice == 'rock':
            print("You lose this turn!", end=" ")
            print("Computer chose {}".format(cpu_choice))
            print("Turn {} ended".format(current_turn))
            current_turn +=1
            loss_counter +=1
            cpu_choice = random.choice(choices)
        if win_counter >= 2:
            print("You won the game!")
            try_again = input("Game Over, try again? ").lower()
            if try_again == "yes":
                current_turn = 0
                win_counter = 0
                loss_counter = 0
                tie_counter = 0
            elif try_again == "no":
                print("Thank you for playing!")
                current_turn = 3
                break
            else:
                print("Please type yes or no.")
        elif loss_counter >= 2:
            print("You lost the game!")
            try_again = input("Game Over, try again? ").lower()
            if try_again == "yes":
                current_turn = 0
                win_counter = 0
                loss_counter = 0
                tie_counter = 0
            elif try_again == "no":
                print("Thank you for playing!")
                current_turn = 3
                break
            else:
                print("Please type yes or no.")
        elif tie_counter == 1 and loss_counter == 1 and win_counter == 1 or tie_counter == 3:
            print("It's a tie!")
            try_again = input("Game Over, try again? ").lower()
            if try_again == "yes":
                current_turn = 0
                win_counter = 0
                loss_counter = 0
                tie_counter = 0
            elif try_again == "no":
                print("Thank you for playing!")
                current_turn = 3
                break
            else:
                print("Please type yes or no.")
    else:
        print("Please be sure to choose/type one of the following: Rock, Paper, Scissors")
```
